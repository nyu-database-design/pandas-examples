# matplotlib examples

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

## Basic style
We will use matplotlib's `seaborn` style.

```python
plt.style.use('seaborn')
# plt.style.available # see all available style options
```

## Simple line plot
Line charts can be easily made from two lists representing the `x` and `y` values.

```python
x1 = [1, 2, 3, 4, 5]
y1 = [10, 4, 20, 15, 13]

plt.plot(x1, y1)
```

Alternatively, the same chart can be created with a dictionary or object containing the labeled data.  We inform the `plot()` function which field to use for x and y

```python
data1 = {
    'seconds': [1, 2, 3, 4, 5],
    'luminescence': [10, 4, 20, 15, 13]
}
plt.plot('seconds', 'luminescence', data=data1)
```

## Further styling
It is possible to further style a plot by supplying an additional argument to the `plot()` function.  This argument will indicate the style and color.

The possible symbols and colors are as follows:
- symbols: `-` , `--`, `-.`, `:` , `.` , `,` , `o` , `^` , `v` , `<` , `>` , `s` , `+` , `x` , `D` , `d` , `1` , `2` , `3` , `4` , `h` , `H` , `p` , `|` , `_`
- colors: `b`, `g`, `r`, `c`, `m`, `y`, `k`, `w`

```python
plt.plot(x1, y1, 'r-.')
```

## Scatter plot
Using the same data, for simplicity, we'll create a scatter plot with the `scatter()` function, passing a few style-related arguments.  See more [documentation on marker options](https://matplotlib.org/stable/api/markers_api.html).

```python
plt.scatter(x1, y1, label='skitscat', color='k', s=25, marker="o")
```

## Simple bar chart
We can make a simple bar chart with the `bar()` function... in this case re-using the same `x` and `y` values.

```python
x2 = [1, 2, 3, 4, 5]
y2 = [4, 14, 16, 1, 12]

plt.bar(x2, y2)
```

## Adding a legend, title, and labels for the axes
To create a legend, give each line a label, and then call the `legend()` function to show them.

```python
plt.plot(x1, y1, label="First bug")

plt.plot(x2, y2, label="Second bug")

# show legend with the labels for each plot
plt.legend() 

# show labels for axes
plt.xlabel('time (seconds)')
plt.ylabel('luminescence')

# create a title for the chart
plt.title('Firefly luminescence over time')
```

## Stacked area chart
A stacked area chart can be created with the `stackplot()` function.  To create labels for the different colored areas, we must create a few blank plots with labels.

```python
# list of 7 days of the week
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# list of suspected cases
suspected = [12, 18, 35, 50, 72, 90, 100]
  
# list of cured cases
cured = [4, 8, 15, 22, 41, 55, 62]
  
# list of deaths
deaths = [1, 3, 5, 7, 9, 11, 13]
  
# plot x-labels, y-label and data
plt.plot([], [], color ='blue', 
         label ='Suspected')
plt.plot([], [], color ='orange',
         label ='Cured')
plt.plot([], [], color ='brown',
         label ='Deaths')
  
# create the stacked area plot
plt.stackplot(days, suspected, cured, 
              deaths, 
              baseline='zero', 
              colors=['blue', 'orange', 'brown'] # use the same colors as the empty plots above
             )

plt.xticks(rotation=45) # rotate x-axis labels by 45 degrees

plt.legend() # add the legend based upon the labels
  
plt.title('No of Cases')
plt.xlabel('Day of the week')
plt.ylabel('Overall cases')
```

## Going further

### numpy
`numpy` is a module for performing optimized vector operations on arrays of *n* dimensions.  Array operation in `numpy` can be up to 50 times faster than operations on regular Python lists.  See [this tutorial](https://www.studytonight.com/numpy/what-is-python-numpy-library) to learn more.

```python
import numpy as np

# create an array of 50 evenly-spaced values from -1 to 1, inclusive.
x = np.linspace(-1, 1, 50) 

# create a classic y = f(x) line
y = 2*x + 1

plt.plot(x, y)
```

### Customizing figures, axes, and plots
The examples above show a fwe of the basic features of `matplotlib` - there are more simple chart types available, such as histograms, pie charts, etc.  

However, in order to dive into more complex visualizations, it is important to master the concepts of figures, axes, and plots - see [this tutorial](https://www.studytonight.com/matplotlib) for more details.

```python
import math
x = np.arange(0, math.pi*2, 0.05)
y = np.cos(x)
fig = plt.figure() # a figure is the canvas upon which all plots are drawn
ax = fig.add_axes([0,0,1,1]) # set the margin of the left and bottom, and the width, and height of the axes compared to the figure
ax.plot(x,y) # create the plot
ax.set_title("cos wave")# used to set the title
ax.set_xlabel('angle')# used to set label for x-axis
ax.set_ylabel('cos') #used to set label for y-axis
```
